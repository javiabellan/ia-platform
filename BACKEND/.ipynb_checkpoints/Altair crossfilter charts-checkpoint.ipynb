```python
import pandas as pd
import altair as alt
from vega_datasets import data
alt.renderers.enable('notebook')
```

```python
source = data.seattle_weather()
source.head()
```

```python
selection_temp    = alt.selection_interval(encodings=['x'])
selection_weather = alt.selection_multi(encodings=['color'])

############################################### QUANTITATIVE: temp_max
hist_temp_max = alt.Chart().mark_bar().properties(width=300, height=200).encode(
    x=alt.X("temp_max:Q", bin=alt.Bin(maxbins=16)),
    y='count()',
    color=alt.condition(selection_temp, alt.value("black"), alt.value("lightgray"))
).add_selection(
    selection_temp
).transform_filter(
    selection_weather
)


############################################### NOMINAL: Weather
bars_weather = alt.Chart().mark_bar().properties(width=300, height=200).encode(
    x='count()',
    y='weather:N',
    color=alt.condition(selection_weather, 'weather:N', alt.value('lightgray'))
).transform_filter(
    selection_temp
).add_selection(
    selection_weather
)



############################################### COMPOSE ALL
alt.vconcat(
    hist_temp_max,
    bars_weather,
    data=source,
    title="Multiview"
)
```

```python
selection_tempMax = alt.selection_interval(encodings=['x'])
selection_precip  = alt.selection_interval(encodings=['x'])
selection_weather = alt.selection_multi(encodings=['y'])


############################################### QUANTITATIVE: temp_max
base = alt.Chart().mark_bar().properties(width=300, height=200).encode(
    x=alt.X("temp_max:Q", bin=alt.Bin(maxbins=16)),
    y='count()'
)

background = base.encode(color=alt.value('lightgray')
).add_selection(selection_tempMax)

highlight = base.encode(color=alt.value('teal')
).transform_filter(selection_tempMax).transform_filter(selection_precip).transform_filter(selection_weather)

# layer the two charts & repeat
hist_tempMax = alt.layer(background, highlight)


############################################### QUANTITATIVE: precipitation
base2 = alt.Chart().mark_bar().properties(width=300, height=200).encode(
    x=alt.X("precipitation:Q", bin=alt.Bin(maxbins=16)),
    y='count()'
)

background2 = base2.encode(color=alt.value('lightgray')
).add_selection(selection_precip)

highlight2 = base2.encode(color=alt.value('teal')
).transform_filter(selection_tempMax).transform_filter(selection_precip).transform_filter(selection_weather)

# layer the two charts & repeat
hist_Preci = alt.layer(background2, highlight2)


############################################### NOMINAL: Weather

base3 = alt.Chart().mark_bar().properties(width=300, height=200).encode(
    x='count()',
    y='weather:N'
)

background3 = base3.encode(color=alt.value('lightgray')
).add_selection(selection_weather)

highlight3 = base3.encode(color='weather:N'
).transform_filter(selection_tempMax).transform_filter(selection_precip).transform_filter(selection_weather)

# layer the two charts & repeat
bars_weather = alt.layer(background3, highlight3)


############################################### TEXT COUNT ROWS

text = base.mark_text().encode(
    text='count()'
)

############################################### COMPOSE ALL
alt.vconcat(
    hist_tempMax,
    hist_Preci,
    bars_weather,
    text,
    data=source,
    title="Multiview"
)
```

```python
data            = source
numericalVars   = ["precipitation", "temp_max", "temp_min", "wind"]
categoricalVars = ["weather"]


def baseHistogram(varName):
    return alt.Chart().mark_bar(
    ).properties(
        width=300,
        height=200,
        title=varName
    ).encode(
        x=alt.X(varName, type='quantitative', bin=alt.Bin(maxbins=16), title=None),
        y=alt.Y('count()', axis=None)
    )

def baseBarchart(varName):
    return alt.Chart().mark_bar(
    ).properties(
        width=300,
        height=200,
        title=varName
    ).encode(
        x=alt.X('count()', axis=None),
        y=alt.Y(varName, type='nominal', title=None),
    )

def crossFilter(charts, filters, varNames, varTypes):
    
    filteredCharts = []

    for (chart, myFilter, varName, varType) in zip(charts, filters, varNames, varTypes):

        # Create the background:
        background = chart.encode(color=alt.value('lightgray')).add_selection(myFilter)

        # Create the filtered highlight:        
        if   varType=="num": highlight = chart.encode(color=alt.value('teal'))
        elif varType=="cat": highlight = chart.encode(color=varName)
        for f in filters:
            highlight = highlight.transform_filter(f)

        filteredCharts.append(alt.layer(background, highlight))
        
    return filteredCharts

    
    
def generateCharts(data, numericalVars, categoricalVars):
    
    numericalCharts   = [baseHistogram(var) for var in numericalVars]
    categoricalCharts = [baseBarchart(var)  for var in categoricalVars]
    
    numericalFilters   = [alt.selection_interval(name=var, encodings=['x']) for var in numericalVars]
    categoricalFilters = [alt.selection_multi(name=var, encodings=['y'])  for var in categoricalVars]

    charts   = numericalCharts  + categoricalCharts
    filters  = numericalFilters + categoricalFilters
    varNames = numericalVars    + categoricalVars
    varTypes = ["num"] * len(numericalCharts) + ["cat"] * len(categoricalCharts)
    
    chartsWithFilters = crossFilter(charts, filters, varNames, varTypes)
    
    return alt.concat(*chartsWithFilters, data=data, columns=2)

    
generateCharts(data, numericalVars, categoricalVars)#.to_dict()
```
