```python
def get_varNames(df): return df.columns.values.tolist()
def get_varTypes(df):
    types = [dtype.name for dtype in df.dtypes.values.tolist()]
    types = [t.replace('object',         'cat')  for t in types]
    types = [t.replace('category',       'cat')  for t in types]
    types = [t.replace('bool',           'cat')  for t in types]
    types = [t.replace('int64',          'num')  for t in types]
    types = [t.replace('float64',        'num')  for t in types]
    types = [t.replace('datetime64[ns]', 'date') for t in types]
    types = [t.replace('timedelta[ns]',  'date') for t in types]
    return types


varNames = get_varNames(data_df)
varTypes = get_varTypes(data_df)

varNames, varTypes
```

```python
def chart_base(varName):
    return alt.Chart().mark_bar().properties(
        width=300, height=200,
        title=varName
    )

def chart_numerical(varName):
    return chart_base(varName).encode(
        x=alt.X(varName, type='quantitative', bin=alt.Bin(maxbins=16), title=None),
        y=alt.Y('count()', axis=None)
    )

def chart_categorical(varName):
    return chart_base(varName).encode(
        x=alt.X('count()', axis=None),
        y=alt.Y(varName, type='nominal', title=None),
    )

def generateCharts(varNames, varTypes):
    charts = []
    for varName, varType in zip(varNames, varTypes):
        if   varType=="num": charts.append(chart_numerical(varName))
        elif varType=="cat": charts.append(chart_categorical(varName))
    return charts

def generateFilters(varNames, varTypes):
    filters = []
    for varName, varType in zip(varNames, varTypes):
        if   varType=="num": filters.append(alt.selection_interval(name=varName, encodings=['x']))
        elif varType=="cat": filters.append(alt.selection_multi(name=varName, encodings=['y']))
    return filters

def generateFilteredCharts(charts, filters, varNames, varTypes):
    filteredCharts = []

    for (chart, myFilter, varName, varType) in zip(charts, filters, varNames, varTypes):

        # Colors
        if varType=="num":
            background = chart.mark_bar(color='teal', opacity=0.15)
            highlight  = chart.mark_bar(color='teal')
            
        if varType=="cat":
            background = chart.mark_bar(color='lightgray')
            highlight  = chart.encode(color=varName + ":N")
        
        # Add my filter to modify others
        background = background.add_selection(myFilter)
        
        # Add others filters that modify me
        for f in filters: highlight = highlight.transform_filter(f)
        
        filteredCharts.append(alt.layer(background, highlight))
        
    return filteredCharts

def univariate_charts(data, varNames, varTypes):
    
    charts         = generateCharts(varNames, varTypes)
    filters        = generateFilters(varNames, varTypes)
    filteredCharts = generateFilteredCharts(charts, filters, varNames, varTypes)
    
    return alt.concat(*filteredCharts, data=data, columns=2)




varNames = ["precipitation", "temp_max", "temp_min", "wind", "weather"]
varTypes = ["num", "num", "num", "num", "cat"]


vega_datasets = [
  "airports", "anscombe", "barley", "burtin", "cars", "crimea",
  "driving", "iowa-electricity", "iris", "la-riots", "seattle-temps",
  "seattle-weather", "sf-temps", "stocks", "us-employment", "wheat"
]

univariate_charts(source_url, varNames, varTypes)
```
